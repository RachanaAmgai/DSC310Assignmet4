```python
# Import necessary libraries
import matplotlib.pyplot as plt
import pandas as pd
import plotly.express as px

# 1. Pie Chart
# Sample data for pie chart
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [15, 30, 45, 10]  # Percentages
colors = ['gold', 'lightcoral', 'lightblue', 'lightgreen']

# Create pie chart
plt.figure(figsize=(7, 7))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart')
plt.axis('equal')  # Equal aspect ratio ensures that pie chart is drawn as a circle.
plt.show()

# 2. Side-by-Side Stacked Bar Chart
# Sample data for side-by-side stacked bar chart
data = {
    'Category': ['A', 'B', 'C', 'D'],
    'Value1': [10, 15, 20, 25],
    'Value2': [20, 25, 30, 35],
    'Value3': [30, 35, 40, 45]
}

df = pd.DataFrame(data)

# Create side-by-side stacked bar chart
df.set_index('Category').plot(kind='bar', stacked=True, figsize=(8, 6))

plt.title('Side-by-Side Stacked Bar Chart')
plt.xlabel('Category')
plt.ylabel('Values')
plt.xticks(rotation=0)
plt.legend(title='Values')
plt.tight_layout()
plt.show()

# 3. Treemap
# Sample data for Treemap
data = {
    'Category': ['A', 'B', 'C', 'D'],
    'Subcategory': ['X', 'Y', 'Z', 'W'],
    'Value': [10, 15, 20, 25]
}

df = pd.DataFrame(data)

# Create Treemap
fig = px.treemap(df, path=['Category', 'Subcategory'], values='Value', title="Treemap Example")
fig.show()
```
